#Challenge 1. Write a Python program to print all permutations of the string "Ironhack"

```python
def permutations(string):
    if len(string) == 1:
        return string
    permutation_list = []
    for x in string:
        for perm in permutations(string.replace(x,'',1)):
            permutation_list.append(x+perm)
    return set(permutation_list)
permutations('ironhack')
```

#Challenge 4: Given an array of distinct integers, return all the possible permutations. 
You can return the answer in any order. 
The lenght of array and the integers are given by user and may vary

```python
def convert_To_Len_th_base(n, arr, Len, L):
    for i in range(L):
        print(arr[n % Len], end = "")
        n //= Len
    print()
def printf(arr, Len, L):
    for i in range(pow(Len, L)):
        convert_To_Len_th_base(i, arr, Len, L)
arr = [1, 2, 3]
Len = len(arr)
L = 3
printf(arr, Len, L)
```
